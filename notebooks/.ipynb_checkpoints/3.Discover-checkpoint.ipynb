### 3.3.4 ------ correlation matrix -------

```python
#for visualization and to prevent any changes to the original copy
train_vis=train.copy()
```

```python
def encode_label(df, col):
    #encode the categories using average salary for each category to replace label
    cat_dict ={}
    cats = df[col].unique().tolist()
    for cat in cats:
        cat_dict[cat] = df[df[col] == cat]['salary'].mean()   
    df[col] = df[col].map(cat_dict)
```

```python
for col in train_vis.columns:
    if train_vis[col].dtype.name == "object":
        encode_label(train_vis, col)
```

```python
fig = plt.figure(figsize=(12, 10))
sns.heatmap(train_vis.corr(), cmap='Blues', annot=True)
plt.xticks(rotation=45)
plt.show()
```

2)Most of "NONE" in "major" is highly related to the "NONE" &"HIGH_SCHOOL".                                                                                                                                                                                      Let's create "Attended_College" and "job_level" to provide more relevance.

```python
def Attended_College(df):
    if(df.degree=="NONE" or df.degree=="HIGH_SCHOOL"):
        return "No"
    else:
        return "Yes"
    
```

```python
train_final['Attended_College'] = train_final.apply(Attended_College, axis=1)
```

```python
def job_level(df):
    if (df.jobType=="JANITOR"): 
        return "JANITOR"

    if((df.jobType=="JUNIOR")or(df.jobType=="SENIOR")):
        return "Average Employees"
    
    if((df.jobType=="MANAGER")or(df.jobType=="VICE_PRESIDENT")):
        return "Management"

    if((df.jobType=="CEO")or(df.jobType=="CTO")or(df.jobType=="CFO")):
        return "Executives"
```

```python
train_final['job_level'] = train_final.apply(job_level, axis=1)
```

>Let's engineer some feature based on the above observations.

###  3.5 ------- Feature Engineering ---------

```python
def feature_workshop(df,feat_list,col_name):
    group=df.groupby(feat_list)
    mean_encode=pd.DataFrame({col_name: group["salary"].mean()})
    df = pd.merge(df, mean_encode, on=feat_list, how='left')
    return df
```

```python
train_final=feature_workshop(train_final,categorical_feat,"cat_feat_mean")
train_final=feature_workshop(train_final,["Experience_bracket","dist_bracket"],"exp_dist_mean")
#train_final=feature_workshop(train_final,["Experience_bracket","job_level","Attended_College"],"exp_job_cllg_mean")
train_final=feature_workshop(train_final,["dist_bracket","job_level"],"dist_job_mean")
```

###  3.5.1  correlation  relation matrix with new features

```python
train_vis=train_final.copy()
for col in train_vis.columns:
    if train_vis[col].dtype.name == "object":
        encode_label(train_vis, col)
```

```python
fig = plt.figure(figsize=(12, 10))
sns.heatmap(train_vis.corr(), cmap='Blues', annot=True)
plt.xticks(rotation=45)
plt.show()
```

```python
#to treat colinearity
train_final.drop(["dist_bracket","job_level","Experience_bracket"],axis=1,inplace=True)
```
